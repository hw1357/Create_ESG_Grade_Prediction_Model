```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
import time

def get_executive_ratios(cdf, my_key, start_year=2019, end_year=2024):
    # 1. API 키 확인
    if not my_key:
        print("🚨 에러: API 키가 없습니다.")
        return None

    # ✅ [수정됨] 올바른 API 엔드포인트 (Ofic 제거)
    base_url = 'https://opendart.fss.or.kr/api/exctvSttus.xml'
    reprt_code = '11011' 

    # --- [데이터 전처리: ID 8자리 강제 변환] ---
    corp_map = cdf[['corp_code', 'corp_name']].copy()
    corp_map['corp_code'] = pd.to_numeric(corp_map['corp_code'], errors='coerce')
    corp_map = corp_map.dropna(subset=['corp_code'])
    corp_map['corp_code'] = corp_map['corp_code'].astype(int)
    corp_map['corp_code'] = corp_map['corp_code'].astype(str).str.zfill(8)
    
    corp_codes = corp_map['corp_code']
    all_year_dfs = []
    
    # 사외이사 판별 키워드
    outside_keywords = ['사외이사', '외부이사', '비상임이사', '기타비상무이사']

    # 봇 차단 방지 헤더
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

    for year in range(start_year, end_year + 1):
        print(f"\n=== {year}년 임원 현황 집계 진행 ===")
        data_records = [] 
        
        for code in tqdm(corp_codes, desc=f"{year}년 기업 진행", ncols=100):
            
            params = {
                'crtfc_key': my_key,
                'corp_code': code,
                'bsns_year': str(year),
                'reprt_code': reprt_code,
            }
            
            # API 호출
            try:
                r = requests.get(base_url, params=params, headers=headers, timeout=10)
                r.raise_for_status()
            except requests.RequestException:
                # 통신 에러 시 건너뜀 -> records에 추가 안 됨 -> 나중에 NaN 처리됨 (조건 1)
                continue

            # 파서 선택
            try:
                soup = BeautifulSoup(r.text, "lxml-xml")
            except:
                soup = BeautifulSoup(r.text, "html.parser")
            
            status = soup.find('status')
            if not status or status.text != '000':
                # 데이터 없음 메시지 등 -> 건너뜀 -> 나중에 NaN 처리됨 (조건 1)
                continue

            executives = soup.find_all('list')
            
            total_count = 0
            female_count = 0
            outside_count = 0
            
            for exe in executives:
                nm = exe.find('nm')
                if not nm or not nm.text.strip():
                    continue
                
                total_count += 1
                
                # -------------------------------------------------
                # 1. 여성 임원 카운트
                # -------------------------------------------------
                sex_tag = exe.find('sexdstn')
                if sex_tag and '여' in sex_tag.text.strip():
                    female_count += 1
                
                # -------------------------------------------------
                # 2. 사외이사 카운트
                # -------------------------------------------------
                ofcps_val = exe.find('ofcps').text.strip() if exe.find('ofcps') else ""
                chrg_val = exe.find('chrg_job').text.strip() if exe.find('chrg_job') else ""
                rgist_val = exe.find('rgist_exctv_at').text.strip() if exe.find('rgist_exctv_at') else ""
                fte_val = exe.find('fte_at').text.strip() if exe.find('fte_at') else ""

                ofcps_clean = ofcps_val.replace(" ", "")
                chrg_clean = chrg_val.replace(" ", "")
                rgist_clean = rgist_val.replace(" ", "")
                fte_clean = fte_val.replace(" ", "")

                is_outside = False
                if any(kw in ofcps_clean for kw in outside_keywords) or \
                   any(kw in chrg_clean for kw in outside_keywords):
                    is_outside = True
                elif '사외이사' in rgist_clean:
                    is_outside = True
                elif '비상근' in fte_clean:
                    is_outside = True
                
                if is_outside:
                    outside_count += 1
            
            # ✅ [로직 수정 및 확인 구간]
            if total_count > 0:
                # 조건 2 & 3: 전체 임원은 있는데, 특정 유형(여성/사외)이 0명이면
                # 0 / total_count = 0.0 으로 정상 계산되어 들어감.
                dir_fe_ratio = (female_count / total_count) * 100
                dir_out_ratio = (outside_count / total_count) * 100
                
                data_records.append({
                    'corp_code': code,
                    'year': year,
                    'DIR_FE': dir_fe_ratio,
                    'DIR_OUT': dir_out_ratio
                })
            else:
                # 데이터가 없거나 total_count가 0인 경우
                # 리스트에 추가하지 않음 -> 아래 merge 단계에서 자동으로 NaN이 됨 (조건 1)
                pass

            time.sleep(0.05) 

        # 연도별 정리
        if data_records:
            year_raw = pd.DataFrame(data_records)
        else:
            # 해당 연도에 아무 데이터도 못 가져왔을 경우 빈 프레임 생성
            year_raw = pd.DataFrame(columns=['corp_code', 'year', 'DIR_FE', 'DIR_OUT'])

        base_df = corp_map.copy()
        base_df['year'] = year
        
        # ✅ Left Merge: 데이터가 없는 기업(records에 없는 기업)은 자동으로 NaN이 됨
        year_df = base_df.merge(year_raw, on=['corp_code', 'year'], how='left')
        all_year_dfs.append(year_df[['corp_name', 'year', 'DIR_FE', 'DIR_OUT']])

    # 최종 병합
    if all_year_dfs:
        final_df = pd.concat(all_year_dfs, ignore_index=True)
    else:
        final_df = pd.DataFrame(columns=['corp_name', 'year', 'DIR_FE', 'DIR_OUT'])

    return final_df
```
