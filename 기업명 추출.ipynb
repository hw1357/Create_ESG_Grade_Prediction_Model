```python
cdf.dropna(subset = 'company', axis = 0, inplace = True)
cdf.dropna(subset = 'stock_code', axis = 0, inplace = True)
```

```python
cdf.loc[cdf['corp_name'] == '신성이엔지', ['stock_code', 'corp_name']]
```

```python
cdf = cdf[cdf['stock_code'] != '021570']
cdf = cdf[cdf['stock_code'] != '003600']
cdf = cdf[cdf['stock_code'] != '000830']
cdf = cdf[cdf['stock_code'] != '037620']
cdf = cdf[cdf['stock_code'] != '053000']
cdf = cdf[cdf['stock_code'] != '002300']
cdf = cdf[cdf['stock_code'] != '104110']
cdf = cdf[cdf['stock_code'] != ' ']
cdf['corp_name'].value_counts()
```
